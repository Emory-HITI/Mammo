```python
# loop
for i in range(10):
    if i == 0:
        continue
    cohort_num = i + 1
    print('running cohort {}'.format(cohort_num))
    original_df = pd.read_csv('/data/mammo/png_kheiron/kheiron_code/cohort_{}/metadata_anon_cohort_{}.csv'.format(cohort_num, cohort_num), low_memory=False)
    original_df = derive_imgType(original_df)
    # identify special views
    original_df = get_spotmag_flags(original_df)
    # change the rooth directories between servers in png_path
    original_df = correct_root_paths(original_df, '/opt/ssd-data/mammo/png_kheiron/', '/data/mammo/png_kheiron/kheiron_code/')
    # removes the old 'png_path' with the 'corrected_png_path' and renames it to 'png_path'
    replace_old_png_path(original_df)
    # make filename and folder_path dictionary for ROI extraction 
    ss_dict = make_screensave_dict(original_df)
    ROI_coords_ssc, ROI_coords_mammo, ROI_matching_ssc, ROI_matching_mammo, ROI_match_levels = ROI.run_extractions(ss_dict, original_df, verbose=False)
    complete_df = pd.DataFrame({'ssc_coord': ROI_coords_ssc, 'ssc_path': ROI_matching_ssc, 'full_coord': ROI_coords_mammo, 'full_path': ROI_matching_mammo, 'levels': ROI_match_levels})
    complete_df.to_pickle('./final_extract/cohort_{}_extract_complete.pkl'.format(cohort_num))
    ssc_df, full_df = generate_map_df(complete_df, verbose=False)
    ssc_df.to_pickle('./final_extract/cohort_{}_extract_SSC.pkl'.format(cohort_num))
    full_df.to_pickle('./final_extract/cohort_{}_extract_FULL.pkl'.format(cohort_num))
    map_df = pd.concat([ssc_df, full_df])
    final_df = map_complete(original_df, map_df)
    final_df.to_pickle('./final_extract/cohort_{}.pkl'.format(cohort_num))
```
